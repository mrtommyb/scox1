```python
%matplotlib inline
from __future__ import division, print_function

import numpy as np
import matplotlib.pyplot as plt
```

```python
import isochrones
from isochrones.dartmouth import Dartmouth_Isochrone
from isochrones import StarModel
dar = Dartmouth_Isochrone(bands=['kep'])
```

```python
#spectroscopic properties (value, uncertainty)
Teff = (3770,128)
logg = (4.74, 0.04)
feh = (-0.05, 0.29)
J = (13.73,0.03)
K = (12.93,0.03)

dar = Dartmouth_Isochrone()

model  = StarModel(dar, Teff=Teff, logg=logg, feh=feh, J=J, K=K)
model.fit_mcmc(nburn=1000,niter=1000,)
```

```python
model.triangle()
```
